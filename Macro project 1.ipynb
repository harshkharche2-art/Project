```python
%gui tk

import tkinter as tk
import random
import time

class MemoryGame:
    def __init__(self, root):
        self.root = root
        self.root.title("Memory Match Game")
        self.moves = 0
        self.first = None
        self.second = None
        self.buttons = []
        self.symbols = list("AABBCCDDEEFFGGHH")
        random.shuffle(self.symbols)

        self.create_widgets()

    def create_widgets(self):
        self.label = tk.Label(self.root, text="Moves: 0", font=("Arial", 16))
        self.label.grid(row=0, column=0, columnspan=4)

        for i in range(4):
            row = []
            for j in range(4):
                btn = tk.Button(self.root, text=" ", width=6, height=3, font=("Arial", 18),
                                command=lambda i=i, j=j: self.reveal(i, j))
                btn.grid(row=i+1, column=j, padx=5, pady=5)
                row.append(btn)
            self.buttons.append(row)

    def reveal(self, i, j):
        if self.buttons[i][j]["text"] == " " and self.second is None:
            self.buttons[i][j]["text"] = self.symbols[i*4 + j]
            if not self.first:
                self.first = (i, j)
            else:
                self.second = (i, j)
                self.root.after(500, self.check)

    def check(self):
        i1, j1 = self.first
        i2, j2 = self.second
        if self.symbols[i1*4 + j1] != self.symbols[i2*4 + j2]:
            self.buttons[i1][j1]["text"] = " "
            self.buttons[i2][j2]["text"] = " "
        self.moves += 1
        self.label.config(text=f"Moves: {self.moves}")
        self.first = None
        self.second = None

        if all(self.buttons[i][j]["text"] != " " for i in range(4) for j in range(4)):
            self.label.config(text=f"Game Over! Total Moves: {self.moves}")

root = tk.Tk()
game = MemoryGame(root)
```
